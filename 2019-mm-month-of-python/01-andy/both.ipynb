```python
import re
import numpy as np
import pandas as pd
import datetime
d = open("data.txt","r").read().rstrip().split("\n")
```

```python
df = pd.concat([pd.DataFrame({k: [v] for k,v in zip(["month", "day", "hour", "minute", "note"], re.match('\[1518-([0-9][0-9])-([0-9][0-9]) ([0-9][0-9]):([0-9][0-9])\] (.*)', x).groups())}) for x in d], ignore_index=True).sort_values(["month", "day", "minute"])
```

```python
df.shape
```

```python
df.head(30)
```

```python
for k in ["month", "day", "hour", "minute"]:
    df[k] = df[k].astype('int')
```

```python
df.tail(30)
```

```python
df['date'] = df.apply(lambda x: datetime.datetime(1518, x.month, x.day, x.hour, x.minute), axis=1)
```

```python
df.loc[df['date'].apply(lambda x: x.hour) == 23, 'date'] = df.loc[df['date'].apply(lambda x: x.hour) == 23, 'date']+datetime.timedelta(hours=1)
```

```python
df.loc[df.hour == 23, 'day'] = df.loc[df.hour == 23, 'day'] + 1
```

```python
df.head(30)
```

```python
df.loc[(df.month==2) & (df.day==29), :]
```

```python
df["Guard"] = df.note.apply(lambda x: "Guard" in x)
```

```python
df["Guard_ID"] = df.loc[df.Guard, 'note'].apply(lambda x: re.match('Guard #([0-9]+).*', x).groups()[0])
```

```python
df.head()
```

```python
df["date_day"] = df.date.apply(lambda x: x.date())
```

```python
df.merge(
    df.groupby('date_day').apply(
        lambda x: x.Guard_ID.values[x.Guard][0]).to_frame().rename(columns={0: 'Guard_IDID'}
    ), 
    on="date_day"
)
```

```python
def f(shift):
    gaurd = shift.Guard_ID.values[shift.Guard][0]
    times = shift.loc[~shift.Guard, :]
    n = int(times.shape[0])
    assert (n % 2) == 0
    # print(times.note.values[list(range(0,n,2))])
    assert (times.note.values[list(range(0,n,2))] == "falls asleep").sum() == (n/2)
    # print(times.note.values[list(range(1,n,2))])
    assert (times.note.values[list(range(1,n,2))] == "wakes up").sum() == (n/2)
    asleep = np.zeros(60)
    for i in range(0,n,2):
        asleep[times.minute.values[i]:times.minute.values[i+1]] += 1
    if n>0:
        assert asleep.sum() > 0
        assert asleep.max() == 1
    return pd.Series({"id": gaurd, "dt": times.shape[0], "asleep": asleep, "total": asleep.sum()})
summed = df.groupby('date_day').apply(f)
summed
```

```python
def h(x):
    mat = np.zeros((int(x.shape[0]),60))
    for i in range(int(x.shape[0])):
        mat[i,:] = x.loc[:,'asleep'].iloc[i]
    return pd.Series({"max": mat.sum(axis=0).max(), 
                      "minute": np.arange(60)[mat.sum(axis=0).max() == mat.sum(axis=0)][0]})
summed.groupby("id").apply(h).sort_values("max")
```
